```python
numbers = [3, 5, 2, 8, 1, 9, 4]
maxofvalue = numbers[0]
for number in numbers:
    if number > maxofvalue:
        maxofvalue = number
print(maxofvalue)
##-----------------------------------------------------
text = "mohamed emad farhat"
vowels = "aeiouAEIOU"
x = 0
for char in text:
    if char in vowels:
        x = x + 1
print(x)
##-----------------------------------------------------
primes = []
for n in range(2, 10):
    for i in range(2, n):
        if n % i == 0:
            break
    else:
        primes.append(n)
print(primes)
##-----------------------------------------------------
n = 10
fib = [0, 1]
for i in range(2, n):
    fib += [fib[i-1] + fib[i-2]]
print(fib)
##-----------------------------------------------------
y = 5000
x = 1200
if x <= y:
    y -= x
    print("balance:", y)
else:
    print("wrong")
##-----------------------------------------------------
list1 = [3, 5, 2, 8, 1, 9, 4]
list2 = [1, 3, 7, 9, 10]
common = []
for x in list1:
    for y in list2:
        if x == y:
            common.append(x)
print(common)
```
